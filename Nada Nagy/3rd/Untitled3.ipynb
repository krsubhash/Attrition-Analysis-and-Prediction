```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
#print the head of the file
df.head()
```

```python
#1. Average_Years_At_Company:

average_Years_At_Company = df['YearsAtCompany'].mean()
print(f"average Years At Company: {average_Years_At_Company:.2f}")
```

```python
# Calculate the average number of years at the company by department
average_years_by_department = df.groupby('Department')['YearsAtCompany'].mean()
print('Average number of years at the company by department:')
print(average_years_by_department)
```

```python
# Plot the average number of years at the company by department
plt.figure(figsize=(10, 6))
average_years_by_department.plot(kind='bar', color='skyblue')
plt.title('Average Number of Years at the Company by Department')
plt.xlabel('Department')
plt.ylabel('Average Number of Years')
plt.xticks(rotation=45)
plt.show()
```

```python
#2.average Years In Current Role
average_Years_In_Current_Role = df['YearsInCurrentRole'].mean()
print(f"average Years In Current Role: {average_Years_In_Current_Role:.2f}")
```

```python
# Calculate the average number of years in the current role by job satisfaction
average_years_by_job_satisfaction= df.groupby('JobSatisfaction')['YearsInCurrentRole'].mean()
print('Average number of years in the current role by job satisfaction:')
print(average_years_by_job_satisfaction)
```

```python
plt.figure(figsize=(10, 6))
average_years_by_job_satisfaction.plot(kind='bar', color='lightgreen')
plt.title('Average Number of Years in Current Role by Performance Rating')
plt.xlabel('Job Satisfaction')
plt.ylabel('Years In Current Role')
plt.xticks(rotation=0)
plt.show()
```

```python
# Calculate the average number of years in the current role by Performance Rating
average_years_by_Performance_Rating= df.groupby('PerformanceRating')['YearsInCurrentRole'].mean()
print('Average number of years in the current role by Performance Rating:')
print(average_years_by_Performance_Rating)
```

```python
# Correlation matrix to see the correlation between years in current role, job satisfaction, and performance rating
correlation_matrix = df[['YearsInCurrentRole', 'JobSatisfaction', 'PerformanceRating']].corr()
print('Correlation matrix:')
print(correlation_matrix)
```

```python
# Plot the correlation matrix
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
#3.average Years Since Last Promotion
average_Years_Since_Last_Promotion = df['YearsSinceLastPromotion'].mean()
print(f"average Years Since Last Promotion: {average_Years_Since_Last_Promotion:.2f}")
```

```python
#4. Examine Distribution of Education Levels:- Explore the distribution of employees' education levels across the organization. Determine if there are any correlations between education level and career advancem
# Define a mapping for education levels if not already clear
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
education_mapping = {
    1: 'Below College',
    2: 'College',
    3: 'Bachelor',
    4: 'Master',
    5: 'Doctor'
}

# Add a readable education level column
df['EducationLevel'] = df['Education'].map(education_mapping)

# Visualize the distribution of education levels
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='EducationLevel', order=['Below College', 'College', 'Bachelor', 'Master', 'Doctor'])
plt.title('Distribution of Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Correlation between education level and career advancement (e.g., JobLevel)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='EducationLevel', y='JobLevel', order=['Below College', 'College', 'Bachelor', 'Master', 'Doctor'])
plt.title('Job Level vs. Education Level')
plt.xlabel('Education Level')
plt.ylabel('Job Level')
plt.xticks(rotation=45)
plt.show()

# Calculate the average JobLevel for each EducationLevel
average_job_level = df.groupby('EducationLevel')['JobLevel'].mean().reset_index()
print(average_job_level)
```
